## Load package and data

```python
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import numpy as np
import seaborn as sns
import sys, os
```

```python
for i, filename in enumerate(os.listdir('data/train/'),1):
    df = pd.read_csv(os.path.join('data/train/',filename))
    if i == 1:
        df_train = pd.DataFrame(columns = list(df.columns).extend(('Well')))
    df['Well'] = f'Well_{i}'
    df_train = pd.concat([df_train,df],axis = 0)
```

```python
sns.pairplot(df_train, hue = 'Well')
```

```python
fig, axs = plt.subplots(1, 9, figsize=(3*9,10), sharey=True)
plt.suptitle(f'Gamma-ray logs of Training data')

for i in range(1,10):
    df = df_train.loc[df_train.Well == f'Well_{i}']   
    axs[i-1].plot(df.GR.values, df.DEPT.values)
    axs[i-1].set_title(f'Well {i}')
    if i == 1:
        axs[i-1].set_ylabel('Depth')
    else:
        axs[i-1].set_ylabel('')
        
    axs[i-1].set_xlabel('Unit: API')
    axs[i-1].grid('on')

plt.gca().invert_yaxis()
```

```python
fig, axs = plt.subplots(1, 9, figsize=(3*9,10), sharey=True)
plt.suptitle(f'Neutron logs of Training data')

for i in range(1,10):
    df = df_train.loc[df_train.Well == f'Well_{i}']   
    axs[i-1].plot(df.NPHI.values, df.DEPT.values)
    axs[i-1].set_title(f'Well {i}')
    if i == 1:
        axs[i-1].set_ylabel('Depth')
    else:
        axs[i-1].set_ylabel('')
    axs[i-1].set_xlabel('Unit: fraction')
    axs[i-1].grid('on')

plt.gca().invert_yaxis()
```

```python
fig, axs = plt.subplots(1, 9, figsize=(3*9,10), sharey=True)
plt.suptitle(f'Density logs of Training data')

for i in range(1,10):
    df = df_train.loc[df_train.Well == f'Well_{i}']   
    axs[i-1].plot(df.RHOB.values, df.DEPT.values)
    axs[i-1].set_title(f'Well {i}')
    if i == 1:
        axs[i-1].set_ylabel('Depth')
    else:
        axs[i-1].set_ylabel('')
        
    axs[i-1].set_xlim([2,2.75])
    axs[i-1].set_xlabel('Unit: g/cc')
    axs[i-1].grid('on')

plt.gca().invert_yaxis()
```

```python
fig, axs = plt.subplots(1, 9, figsize=(3*9,10), sharey=True)
plt.suptitle(f'Resistivity logs of Training data')

for i in range(1,10):
    df = df_train.loc[df_train.Well == f'Well_{i}']   
    axs[i-1].plot(df.RD.values, df.DEPT.values)
    axs[i-1].set_title(f'Well {i}')
    if i == 1:
        axs[i-1].set_ylabel('Depth')
    else:
        axs[i-1].set_ylabel('')
        
    axs[i-1].set_xlim([0,25])
    axs[i-1].set_xlabel('Unit: Ωm')
    axs[i-1].grid('on')

plt.gca().invert_yaxis()
```
