```python
# Now we are going to visualize Histogram of GR in training data #1

plt.figure(figsize = (3,3))

plt.hist(df.GR.values)
plt.xlabel('GR')
plt.ylabel('Frequency')
plt.title('Gamma-ray record in Train #1')
plt.grid('on')

```

```python

plt.figure(figsize = (3,10))

plt.plot(df.GR.values, df.DEPT.values)

plt.gca().invert_yaxis()
plt.title('GR')
plt.ylabel('Depth')
plt.xlabel('Unit: API')
plt.grid('on')
```

```python
df.columns
```

```python
plt.figure(figsize = (12,10))
plt.suptitle('Training data #1')
### 첫번째 Subplot
plt.subplot(1,4,1)
plt.plot(df.GR.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('Gamma-ray')
plt.ylabel('Depth')
plt.xlabel('Unit: API')
plt.grid('on')

### 두번째 Subplot
plt.subplot(1,4,2)
plt.plot(df.RHOB.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('Density')
plt.ylabel('Depth')
plt.xlabel('Unit: g/cc')
plt.xlim([2,2.75])
plt.grid('on')

### 세번쨰 Subplot
plt.subplot(1,4,3)
plt.plot(df.NPHI.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('Neutron')
plt.ylabel('Depth')
plt.xlabel('Unit: %')
plt.grid('on')

### 네번째 Subplot
plt.subplot(1,4,4)
plt.plot(df.RD.values, df.DEPT.values)
plt.gca().invert_yaxis()
plt.title('Resistivity')
plt.ylabel('Depth')
plt.xlabel('Unit: Ωm')
plt.xlim([0,25])
plt.grid('on')

plt.tight_layout()

plt.savefig('WellLogs_Training#1.png')
```

```python
for i in range(1,10):
    plt.figure(figsize = (12,10))
    plt.suptitle(f'Training data # {i}')
    df = pd.read_csv(f"data/train/aligned_well_0{i}.csv")   
    
    ### 첫번째 Subplot
    plt.subplot(1,4,1)
    plt.plot(df.GR.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('Gamma-ray')
    plt.ylabel('Depth')
    plt.xlabel('Unit: API')
    plt.grid('on')

    ### 두번째 Subplot
    plt.subplot(1,4,2)
    plt.plot(df.RHOB.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('Density')
    plt.ylabel('Depth')
    plt.xlabel('Unit: g/cc')
    plt.xlim([2,2.75])
    plt.grid('on')

    ### 세번쨰 Subplot
    plt.subplot(1,4,3)
    plt.plot(df.NPHI.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('Neutron')
    plt.ylabel('Depth')
    plt.xlabel('Unit: %')
    plt.grid('on')

    ### 네번째 Subplot
    plt.subplot(1,4,4)
    plt.plot(df.RD.values, df.DEPT.values)
    plt.gca().invert_yaxis()
    plt.title('Resistivity')
    plt.ylabel('Depth')
    plt.xlabel('Unit: Ωm')
    plt.xlim([0,25])
    plt.grid('on')

    plt.tight_layout()

    plt.savefig(f'WellLogs_Training#{i}.png')
```
